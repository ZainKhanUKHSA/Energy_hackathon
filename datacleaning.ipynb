```python
import pandas as pd

# File path
file_path_electricity = 'data/LSOA_domestic_elec_2010-22.xlsx'
file_path_gas = 'data/LSOA_domestic_gas_2010-2022.xlsx'

# Load the specific sheet and skip unnecessary rows
data_elec_2022 = pd.read_excel(file_path_electricity, sheet_name='2022', skiprows=4)
data_gas_2022 = pd.read_excel(file_path_gas, sheet_name= '2022', skiprows=4)


```

```python
print(data_elec_2022.columns)
```

```python
# Rename columns by adding "Electricity" before the existing names
data_elec_2022.rename(columns={
    'Number\nof meters': 'Electricity Number of Meters',
    'Total \nconsumption\n(kWh)': 'Electricity Total Consumption (kWh)',
    'Mean \nconsumption\n(kWh per meter)': 'Electricity Mean Consumption (kWh per Meter)',
    'Median \nconsumption\n(kWh per meter)': 'Electricity Median Consumption (kWh per Meter)'
}, inplace=True)

# Display updated column names
print(data_elec_2022.columns)
```

```python
data_elec_2022.head()
```

```python
data_gas_2022.head()
```

```python
# Rename columns by adding "Gas" before the existing names
data_gas_2022.rename(columns={
    'Number\nof meters': 'Gas Number of Meters',
    'Total \nconsumption\n(kWh)': 'Gas Total Consumption (kWh)',
    'Mean \nconsumption\n(kWh per meter)': 'Gas Mean Consumption (kWh per Meter)',
    'Median \nconsumption\n(kWh per meter)': 'Gas Median Consumption (kWh per Meter)',
    'Number of\nnon-consuming meters': 'Gas Number of Non-Consuming Meters'
}, inplace=True)

# Display updated column names
print(data_gas_2022.columns)
```

```python
data_gas_2022.head()
```

```python
joined_data = pd.merge(data_gas_2022, data_elec_2022, on='LSOA code', how='inner')
```

```python
joined_data.head()
```

```python
joined_data.columns
```

```python
# List of columns to drop
columns_to_drop = [
    'Local authority code_x',
    'Local authority_x',
    'MSOA code_x',
    'Middle layer super output area_x',
    'Lower layer super output area_x',
    'Local authority code_y',
    'Local authority_y',
    'MSOA code_y',
    'Middle layer super output area_y',
    'Lower layer super output area_y'
]

# Drop the columns from the merged dataset
joined_data = joined_data.drop(columns=columns_to_drop)

# Display the first few rows to verify
print(joined_data.head())
```

```python
joined_data.head()
```

```python
import pandas as pd

# File path
file_path = 'data/gdf_export.xlsx'  # Adjust the path if the file is not in the current directory

# Read the Excel file into a DataFrame
lsoa = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
print(lsoa.head())
```

```python
lsoa.head()
```

```python
lsoa = lsoa.drop(columns=['type', 'id'])
```

```python
lsoa.head()
```

```python
# Merge the two DataFrames on the 'LSOA code' column
final_data = pd.merge(joined_data, lsoa, on='LSOA code', how='inner')


```

```python
final_data.head()
```

```python
# Split the 'centroid' column to extract longitude and latitude
final_data[['longitude', 'latitude']] = final_data['centroid'].str.extract(r'POINT \(([-\d.]+) ([-\d.]+)\)')

# Convert them to numeric for further analysis
final_data['longitude'] = pd.to_numeric(final_data['longitude'])
final_data['latitude'] = pd.to_numeric(final_data['latitude'])

# Drop the original 'centroid' column if it's no longer needed
final_data = final_data.drop(columns=['centroid'])
```

```python
# Save the DataFrame to an Excel file
final_data.to_excel('final_data.xlsx', index=False)


```
